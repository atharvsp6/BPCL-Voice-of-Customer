```python
import pandas as pd 

df = pd.read_csv("raw_reviews.csv")
df.head()
```

```python
df = df[['content','score','at']]
df.rename(columns={'at':'date'}, inplace=True)
df['date'] = pd.to_datetime(df['date'])
df.dropna(inplace=True)
df.info()
```

```python
import re

def clean_reviews(text):
    text = str(text).lower()
    text = re.sub(r'[^a-z0-9\s]', '', text)
    return text

df['clean_text'] = df['content'].apply(clean_reviews)

df[['content', 'clean_text']].head()
```
